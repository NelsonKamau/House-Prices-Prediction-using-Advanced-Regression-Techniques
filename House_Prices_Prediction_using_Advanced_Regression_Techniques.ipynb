## **4. Data Cleaning**

```python
# Finding Missing Percentage

def missing_percentage(X):
  missing = X.isnull().sum()
  missing = missing[missing > 0]
  missing_percentage = round(missing / len(X), 3) * 100
  print(pd.DataFrame({'Number of missing values': missing, 'Percentage': missing_percentage}).sort_values(by = 'Percentage', ascending=False))
```

```python
missing_percentage(train)
```

```python
missing_percentage(test)
```

```python
Y_train = train['SalePrice']
train.drop(columns=['SalePrice'], inplace=True)
```

```python
df = pd.concat([train, test])
df.head()
```

```python
df.drop(['PoolQC', 'MiscFeature', 'Alley', 'Fence', 'FireplaceQu', 'Id'], axis=1, inplace=True)
df.head()
```

```python
new_categorical = df.select_dtypes('object').columns
new_categorical
```

```python
for col in df.columns:
  if col in new_categorical:
    mode = df[col].mode()[0]
    df[col] = df[col].fillna(mode)
  else:
    median = df[col].median()
    df[col] = df[col].fillna(median)
```

```python
missing_percentage(df)
```

```python
df = pd.get_dummies(df, new_categorical)
```

```python
train_num = train.shape[0]
X_train = df[:train_num]
X_test =df[train_num:]
print(f'X_train shape: {X_train.shape}')
print(f'X_test shape: {X_test.shape}')
```

## **5. Model Building**


```python
XGB = XGBRegressor()
param = {'max_depth':[5, 10], 'booster':['gbtree','bg linear']}
cv = 5
```

```python
model_XGB = GridSearchCV(XGB, param, n_jobs=-1, cv=cv)
model_XGB.fit(X_train, Y_train)
```

```python
ypred = model_XGB.predict(X_test)
```

## **6. Make Submission**

```python
submission = pd.DataFrame({
    'Id': test['Id'],
    'SalePrice': ypred
})
```
